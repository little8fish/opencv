```python
def show_img(img, mode=None, figsize=None):
    if figsize:
        plt.figure(figsize=figsize)
    plt.imshow(img[:,:,::-1], cmap=mode)
    plt.axis('off')
    plt.show()
```

```python
img1 = cv.imread('/Users/dengjiacheng/PycharmProjects/opencv1/normal/image/view.jpg')
```

```python
show_img(img1)
```

```python
img2 = cv.imread('/Users/dengjiacheng/PycharmProjects/opencv1/normal/image/rain.jpg')
```

```python
show_img(img2)
```

```python
img3 = cv.add(img2, img1)
```

```python
show_img(img3)
```

```python
img4 = img2+img1
```

```python
show_img(img4)
```
